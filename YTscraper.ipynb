```python
art_dict = {'Video ID':[], 'Title':[],'Description':[], 'Category':[]}
```

```python
ctr = 0
```

```python
def grabVideos(keyword,token=None):
    res = youtube_search(keyword)
    token = res[0]
    videos = res[1]
    for vid in videos:
        art_dict['Video ID'].append(vid['id']['videoId'])
        art_dict['Title'].append(vid['snippet']['title'])
        art_dict['Description'].append(vid['snippet']['description'])
        art_dict['Category'].append('art')
    print("added " + str(len(videos)) + " videos to a total of " + str(len(art_dict['Video ID'])))
    return token
```

```python
token = grabVideos("art and music")
while ctr<=36:
    token = grabVideos("art and music", token=token)
    ctr+=1
```

```python
with open("art.csv", "w") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(art_dict.keys())
    writer.writerows(zip(*art_dict.values()))
```
